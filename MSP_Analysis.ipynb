```python
# Pearson Correlation Coefficient helps to understand the relationship between 2 variables,stands always between -1 to 1.
from scipy.stats.stats import pearsonr
pcc = pearsonr(November['EarlyHours'],November['Hours'])
print(pcc)
```

```python
# To analyse the relationship between variables, scatter plot helps.
plt.figure(figsize=(13,8))
plt.plot(November['EarlyHours'], November['Hours'], marker='.', linestyle='None')
plt.xlabel('Early Hours Login')
plt.ylabel('Total working Hours')
plt.title('Relation betwen early login time and Total working Hours')
plt.tight_layout()
plt.show()
```

```python
def pearson_r(x,y):
    corr_mat = np.corrcoef(x,y)
    return corr_mat
novem = November.dropna(how='any')
pcc = pearson_r(novem['EarlyHours'],novem['Hours'])
print(pcc)
```

```python
# Employees working for n no. of hours per day is exponentially distributed. ie we don't know login an login time of each 
# employee and they are independent of each other.
plt.figure(figsize=(13,8))
from random import random
x = np.sort(November['Hours'])
y = np.arange(1, len(x)+1) / len(x)

mean = November['Hours'].mean()
samples = np.random.exponential(mean, size=10000)
x_theor = np.sort(samples)
y_theor = np.arange(1, len(samples)+1) / len(samples)

plt.plot(x_theor, y_theor)
plt.plot(x, y, marker='.', linestyle='none',color='r')
plt.xlabel('No. Of Hours')
plt.ylabel('ECDF')
plt.margins(0.02)
plt.show()
```
